# Analysis of the distribution of volume

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the CSV
df = pd.read_csv("../5. Robust Estimation and Evaluation Methods/analysis/droplet_analysis.csv")  # Replace with your actual file path

# Drop rows where Volume is missing (just in case)
df = df.dropna(subset=["Volume"])

# Show basic stats
min_volume = df["Volume"].min()
max_volume = df["Volume"].max()
mean_volume = df["Volume"].mean()
std_volume = df["Volume"].std()

print(f"Volume Range: {min_volume:.2f} to {max_volume:.2f}")
print(f"Mean Volume: {mean_volume:.2f}")
print(f"Std Dev: {std_volume:.2f}")

# Histogram
plt.figure(figsize=(8, 5))
plt.hist(df["Volume"], bins=30, color="skyblue", edgecolor="black")
plt.title("Distribution of Volume")
plt.xlabel("Volume")
plt.ylabel("Frequency")
plt.grid(True)
plt.tight_layout()
plt.show()
```
